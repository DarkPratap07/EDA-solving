```python
# Use pandas qcut to automatically create 5 equal-sized age groups
df['Age_Group'] = pd.qcut(df['Age'], 5)

# Calculate average spending score for each group
age_spending = df.groupby('Age_Group')['Spending_Score'].mean()
print(age_spending)

# Optional: Bar plot
age_spending.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Average Spending Score by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Spending Score")
plt.show()
```
